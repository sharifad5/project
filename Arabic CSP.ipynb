# CSP

## مقدمة

تحتوي مشكلة الرضا القيد (CSP) على مجموعة من المتغيرات، ومجال من القيم لكل متغير ومجموعة من القيود. الهدف الرئيسي هو جعل كل متغير له قيمة بحيث يتم استيفاء جميع القيود.
لتسهيل الأمر ، لنفترض أننا ننظر إلى خريطة أستراليا التي تُظهر كل ولاية من ولاياتها ، كما هو موضح في الشكل أدناه ، وأننا مُكلفون بمهمة تلوين كل منطقة إما باللون الأحمر أو الأخضر أو ​​الأزرق بطريقة لا توجد بها دول مجاورة المناطق لها نفس اللون. لصياغة هذا باعتباره CSP ، نحدد المتغيرات لتكون المناطق: WA و NT و Q و NSW و V و SA و T. مجال كل متغير هو المجموعة {أحمر؛ أخضر؛ أزرق}. تتطلب القيود أن يكون للمناطق المجاورة ألوان مميزة ؛ على سبيل المثال ، المجموعات المسموح بها لـ WA و NT هي الأزواج
{(أحمر ، أخضر) ، (أحمر ، أزرق) ، (أخضر ، أحمر) ، (أخضر ، أزرق) ، (أزرق ، أحمر) ، (أزرق ، أخضر)}.
(يمكن أيضًا تمثيل القيد بشكل أكثر إيجازًا على أنه عدم المساواة WA ≠ NT ، بشرط أن يكون لخوارزمية رضا القيد طريقة ما لتقييم مثل هذه التعبيرات.)
هناك العديد من الحلول الممكنة ، مثل {WA = أحمر ، NT = أخضر ، Q = أحمر ، NSW = أخضر ، V = أحمر ، SA = أزرق ، T = أحمر}.



هناك أيضًا آليات لنمذجة القيود أو الشروط المطلوبة لتحقيق أو تحقيق هدف معين ، لذلك يتم حل (تقنيات الرضا القيد) أو (مشكلات الرضا القيد) (CSP) من خلال خوارزميات البحث التي تندرج تحت الذكاء الاصطناعي ، أمثلة على هذه المشاكل:

- حل لعبة سودوكو
- مشكلة الملكات

سوف نتعلم المزيد عن مشاكل الرضا عن القيود وكيفية حلها باستخدام خوارزميات البحث باستخدام لغة بايثون.

## مشكلة الملكات
مشكلة الملكات تتمثل في انه يوجد لوح شطرنج بمساحة ن * ن و يجب وضع عدد ن من الملكات بحيث لا تتعدى اي ملكة على الاخرى، و من المعلوم ان الملكة بأمكانها التحرك بشكل عمودي او افقي او قطري.



هذا هو شكل اللوح بعد حل المشكلة



```python
N = 4
```

```python
board = [[0]*N for _ in range(N)]
```

```python
"function to print the board"
def printBoard(board,N): 
    for x in range(N):
        for y in range(N):
            print(board[x][y], end = " ")
        print() 
```

```python
printBoard(board,N)
```

# مقارنة الخوارزميات


لإجراء مقارنة بسيطة بين خوارزميتين ، دعنا نعلن عن عداد لحساب عدد التحقق من الحلقات الداخلية

```python
class sol:
    bk = 0 # to count loops Backtraking
    fc = 0 # to count loops Forwared checking
    def is_attack(i, j):
        for k in range(0,N):
            if board[i][k] == 1 or board[k][j] == 1:
                return True
        for k in range(0,N):
            for l in range(0,N):
                if (k+l == i+j) or (k-l == i-j):
                    if board[k][l] == 1:
                        #print('FC count') # when printed this it count a loop in the total loops for Forward Checking algrithm
                        sol.fc += 1
                        return True
        return False

    def N_queen(n):
        if n == 0:
            return True
        for i in range(0,N):
            for j in range(0,N):
                if (not(sol.is_attack(i,j))) and (board[i][j]!=1):
                    board[i][j] = 1
                    sol.bk = sol.bk + 1
                    #print('BK count') # when printed this it count a loop in the total loops for Bactraking algrithm
                    if sol.N_queen(n-1)==True:
                        return True
                    board[i][j] = 0          
        return False


N = 4 
board = [[0]*N for _ in range(N)] 

sol.N_queen(N)
print('N = 4')
print(sol.bk)
print(sol.fc)

#---------------------------
N = 6 
board = [[0]*N for _ in range(N)] 

sol.N_queen(N)
print('N = 6')
print(sol.bk)
print(sol.fc)
#---------------------------
N = 7 
board = [[0]*N for _ in range(N)] 

sol.N_queen(N)
print('N = 7')
print(sol.bk)
print(sol.fc)
#---------------------------
N = 8 
board = [[0]*N for _ in range(N)] 

sol.N_queen(N)
print('N = 8')
print(sol.bk)
print(sol.fc)
#---------------------------
N = 10 
board = [[0]*N for _ in range(N)] 

sol.N_queen(N)
print('N = 10')
print(sol.bk)
print(sol.fc)
```
